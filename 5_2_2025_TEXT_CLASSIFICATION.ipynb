```python
#Generate word frequency
def gen_freq(text):
  word_list = []
  for tw_words in text.split():
    word_list.extend(tw_words)
  word_freq = pd.Series(word_list).value_counts()
  word_freq = word_freq.drop(stop, errors='ignore')
  return word_freq
```

```python
#Check whether a negation term is present in the text
def any_neg(words):
  for word in words:
    if word in ['n', 'no', 'non', 'not'] or re.search(r"\wn't", word):
      return 1
    else:
      return 0
```

```python
#Check whether one of the 100 word rare words is present in the text
def any_rare(words, rare_100):
  for word in words:
    if word in rare_100:
      return 1
    else:
      return 0
```

```python
#Check whether prompt words are present
def is_question(words):
  for word in words:
    if word in ['when', 'what', 'how', 'why', 'who', 'where']:
      return 1
    else:
      return 0
```

```python
!pip install scikit-learn
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier

# Assuming 'dataset' is your DataFrame and 'tweet' is the column with text data
# and 'label' is the column with your target variable

# 1. Create a TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# 2. Fit the vectorizer to your text data and transform it
X = vectorizer.fit_transform(dataset['tweet'])

# 3. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, dataset['label'], test_size=0.2, random_state=42) # Adjust test_size and random_state as needed

# 4. Create and train the RandomForestClassifier
clf_rf = RandomForestClassifier()
clf_rf.fit(X_train, y_train)

# 5. Make predictions
rf_pred = clf_rf.predict(X_test).astype(int)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_rf=RandomForestClassifier()
clf_rf.fit(X_train,y_train)
rf_pred=clf_rf.predict(X_test).astype(int)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
print("Accuracy:", accuracy_score(y_test, rf_pred))
```
